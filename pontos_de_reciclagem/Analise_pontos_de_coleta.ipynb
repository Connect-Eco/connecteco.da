```python
caracteres_anormais = [chr(146), chr(147), chr(148), chr(150)]

def deletar_caracteres_anormais(dataset, caracteres):
    for coluna in dataset.columns:
        for caracter in caracteres:
            dataset.loc[:, coluna] = dataset[coluna].apply(lambda x: str(x).replace(caracter, ''))
    return dataset

dataset_final = deletar_caracteres_anormais(dataset_final, caracteres_anormais)
```

# Visualizando dados

```python
estado_counts = dataset_final['estado'].value_counts()

top_5_estados = estado_counts.head(5)
other_estados_count = estado_counts[5:].sum()

top_5_estados['Outros'] = other_estados_count

labels = [f'{state}' for state in top_5_estados.index]

plt.figure(figsize=(12, 6), dpi = 300)
plt.pie(
    top_5_estados,
    labels=labels,
    startangle=140,
    labeldistance=1.05,  
    wedgeprops={'edgecolor': 'black'})
plt.title('Distribuição de Informações Coletadas por Estado')
plt.axis('equal') 
plt.show()
```

```python
material_counts = dataset_final['tipo_de_material'].value_counts()

plt.figure(figsize=(14, 8), dpi = 300)
bars = plt.bar(material_counts.index, material_counts, color=plt.cm.Paired(range(len(material_counts))), edgecolor='black')

plt.title('Distribuição dos Tipos de Material Coletados')
plt.xlabel('Tipo de Material')
plt.ylabel('Frequência')

plt.xticks(rotation=45, ha='right')

plt.tight_layout() 
plt.show()
```

```python
tipo_counts = dataset_final['tipo'].value_counts()

plt.figure(figsize=(10, 6), dpi = 500)
bars = plt.bar(tipo_counts.index, tipo_counts, color=plt.cm.Paired(range(len(tipo_counts))), edgecolor='black')

plt.title('Tipos de filiação dos pontos de coleta')
plt.xlabel('Tipo')
plt.ylabel('Frequência')

plt.xticks(rotation=45, ha='right')

plt.tight_layout() 
plt.show()
```
