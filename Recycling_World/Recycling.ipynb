Agora irei utilizar uma função para visualizar a distribuição dos dados em histogramas.

```python
def graph_hist(df, valor):
    # Selecionar apenas as colunas numéricas
    numeric_columns = df.iloc[:, valor].select_dtypes(include=['number'])

    # Calcular o número de linhas e colunas
    num_plots = len(numeric_columns.columns)
    num_cols = 3
    num_rows = (num_plots + num_cols - 1) // num_cols  # Calcula o número de linhas necessário

    # Criar uma figura e eixos
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, num_rows * 4))

    # Iterar sobre as colunas numéricas e gerar histogramas
    for i, column in enumerate(numeric_columns.columns):
        row = i // num_cols
        col = i % num_cols
        sns.histplot(df[column], ax=axes[row, col], kde=False)
        axes[row, col].set_title(column, fontsize=10)
        
    # Remover eixos vazios, se houver
    for j in range(num_plots, num_rows * num_cols):
        fig.delaxes(axes.flatten()[j])

    # Ajustar o layout
    plt.tight_layout()
    plt.show()
```

```python
graph_hist(country,  range(1, 14))
```

- **Madeira**: Representa e média apenas 5 % do lixo.
- **Jardinagem**: Varia de 1% a 15% do lixo.

## Brasil Comparado ao Mundo

Farei uma comparação com os dados do Brasil em relação a **mediana** do mundo. A mediana foi escolhida pois a distribuição possui muitos pontos extremos.

```python
# Calcula a mediana de cada coluna no DataFrame 'country'
world_median = country.median()

# Transforma a série resultante da mediana em um DataFrame de uma linha
world_median = pd.DataFrame(world_median).transpose()

# Adiciona as informações do Brasil ao DataFrame 'world_median'
world_median = world_median.append(brazil)

# Adiciona uma coluna 'country_name' com os nomes 'World' e 'Brazil' para identificar as linhas
world_median['country_name'] = ['World', 'Brazil'] 

# Exibe o DataFrame resultante
world_median
```

```python
def graph_axis(df, valor, name, ylabel,xlabel):
    # Selecionar apenas as colunas numéricas
    numeric_columns = df.iloc[:, valor].select_dtypes(include=['number'])

    # Calcular o número de linhas e colunas
    num_plots = len(numeric_columns.columns)
    num_cols = 3
    num_rows = (num_plots + num_cols - 1) // num_cols  # Calcula o número de linhas necessário

    # Criar uma figura e eixos
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, num_rows * 4))

    # Iterar sobre as colunas numéricas e gerar gráficos de barras
    for i, column in enumerate(numeric_columns.columns):
        row = i // num_cols
        col = i % num_cols
        sns.barplot(x=df[name], y=df[column], ax=axes[row, col])
        axes[row, col].set_title(column,fontsize=10)
        axes[row, col].set_xlabel(xlabel,fontsize=8)
        axes[row, col].set_ylabel(ylabel,fontsize=8)
        axes[row, col].set_xticklabels(df[name], rotation=40, fontsize=8)

    # Remover eixos vazios, se houver
    for j in range(num_plots, num_rows * num_cols):
        fig.delaxes(axes.flatten()[j])

    # Ajustar o layout
    plt.tight_layout()
    plt.show()
```

```python
graph_axis( world_median, range(1, 10),'country_name', 'Porcentagem(%)', 'País')
```

Comparando o Brasil com o Mundo:

- O Brasil produz proporcionalmente **menos** resíduos de vidro.

- O Brasil produz proporcionalmente **igual ou próximo** resíduos de metal, papel, plástico e outros.

- O Brasil produz proporcionalmente **mais** resíduos orgânicos

- O dataframe não possui dados do Brasil para borracha, madeira e jardinagem.

Através dessa visualização os tipos de resíduos do Brasil são semelhantes ao do mundo.

```python
graph_axis(world_median, range(11, 15),'country_name', 'Toneladas', 'País')
```

O Brasil produz em toneladas uma quantidade muito maior de resíduos. Devemos levar em cosideração que a maior parte dos píses são pequenos e com populações pequenas, diferente do Brasil.

```python
graph_axis(world_median, range(18, 26),'country_name', 'Porcentagem', 'País')
```

- O Brasil possui uma **porcentagem bem menor de aterros abertos do que o resto do mundo**, no Brasil representa 15% enquanto no mundo a mediana é de 60%.

- O Brasil possui uma **porcentagem maior de aterros com incineração em relação ao mundo**, mas possui uma porcentagem menor de aterros fechados.

- EM relação a reciclagem o Brasil está muito átras em relação ao mundo, **reciclando apenas 1.4% enquanto a mediana do mundo é de 14%, 10 vezes mais que o Brasil**.

- Plástico representa em média 12% dos resíduos, variando entre 5% e 25%.

## Tratamento de Resíduos

- **Os aterros são o método mais comum de tratamento de resíduos**.
- Há uma grande variação no destino dos resíduos entre diferentes países, com reciclagem variando de 0% a 67%, e uma mediana global de 5%.
- Aterros abertos são comuns mundialmente, mas muitos países utilizam aterros fechados com incineração.

## Comparação do Brasil com o Mundo

- **O Brasil possui uma porcentagem menor de aterros abertos (15%) comparado com a mediana mundial (60%)**.
- **O Brasil recicla apenas 1,4% dos resíduos, muito abaixo da mediana mundial de 14%**.
- Resíduos orgânicos nas cidades brasileiras são semelhantes à média global.
- Cidades brasileiras, como Brasília, apresentam mais resíduo de papel que a média global.
- Cidades brasileiras, como o Rio de Janeiro, têm mais resíduo de plástico que a média global.

## Considerações Finais

O Brasil possui várias características de composição de resíduos e tratamento desses semelhantes a média do mundo, com algumas diferenças, o mesmo se aplica as principais cidades do Brasil e as outras cidades do mundo. Entre as diferenças está a melhor qualidade dos aterros, por possuirem incineradoras, e uma quantidade muito menor de aterros abertos, que são muito prejudiciais ao meio ambiente. Uma diferença preocupante foi a que o Brasil recicla apenas 1.4% do lixo, 40 vezes menor que a média do mundo de 14%. Sendo assim, essa análise demonstra que o **Brasil necessita de projetos para o aumento da reciclagem, que está em situação muito pior em relação ao mundo**.
